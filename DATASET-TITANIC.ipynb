# Aplicação de filtros multiplos


```python
'''Na pesquisa agora, será solicitado a partir de um filto, que seja exibido somente os passageiros
do sexo feminino e com idade maior ou igual a 30. 
Para definir duas condições e que as duas precisam ser cumpridas, devemos o & que significa and.'''
df[(df ['Sex'] == 'female') & (df['Age'] >= 30)]
```

```python
'''Podemos usar a função loc para executar a mesma pesquisa que a anterior, utilizando o rotulo da coluna 
e uma ou mais condições dentro das colunas desejadas. Alem do & para que só seja retornado resultdados 
que cumpram todas as condições utilizadas.'''
df.loc[(df ['Sex'] == 'female') & (df['Age'] >= 30)]
```

```python
'''Diferente do operador And, podemos utilizar o operado or que  é representado pelo |, para botar duas ou mais 
condições e o resultado do filtro será composto por todos os dados que cumprirem uma ou outra condição, também podendo
cumprir todas as condições pesquisadas.

Nesse filtro iremos solicitar o retorno de todos dos dados em que um passageiro tem idade maior ou igual a 30 anos
ou que seja do sexo feminino.'''
df[(df['Age'] >= 30) | (df['Sex'] =='female')]
```

```python
'''Fazendo um filtro com três condições, tais quais todas precisam ser cumpridas.'''

df[(df['Sex'] == 'male') & (df['Age'] >=30) & (df['Embarked'] == 'S')]
```

```python
'''Utilizando a função index e a função to_list, podemos crir uma lista contendo somente o valor do index das informações
que cumpriram as exigências do filtro executado.'''

df[(df['Sex'] == 'male') & (df['Age'] >= 30) & (df['Embarked'] == 'S')].index.to_list()
```

```python
'''Verificando as métricas com agrupamento por sexo, em relação a idade e valor de ticket pago.'''
variaveis = ['Age', 'Fare'] #colunas que serão usadas
group1 = (df.groupby(by='Sex')[variaveis]) #Com o groupby, farei o agrupamento das colunas acima pelo gênero.
metricas = [np.mean, np.std, np.median, min, max] #Valores que serão obtidos, media, desvio padrão,mediana, minima e máxima
```

```python
'''Retornando as métricas desejadas.'''
group1.agg(metricas)
'''Verificamos que em média, os homens presentes no Titanic tinham idade maior que as mulheres e 
que as mulheres pagaram um valor médio de ticket maior.'''
```

```python
'''Verificando as métricas com agrupamento por sexo e por sobreviventes ou não, em relação a idade
e valor de ticket pago. '''
variaveis = ['Age', 'Fare']
group2 = df.groupby(by= ['Sex', 'Survived'])[variaveis] 
```

```python
'''Retornando as métricas desejadas.'''
group2.agg(metricas).T
'''Verificamos que em média, os homens presentes no Titanic tinham idade maior que as mulheres e 
que as mulheres pagaram um valor médio de ticket maior.
Também obtivemos que as muheres que não sobreviveram tinham em média 25.04 anos e as que sobreviveram tinha em 
média 28.84 anos. Com relação aos Homens, podemos ver que os que não sobreviveram, tinham 31,61 anos e os que 
sobreviveram tinham 27.27 anos.''
```

```python
'''Imprimindo um mapa da correlação'''
sns.heatmap(corr)
```

```python
'''Criando uma mascara para poder plotar um gráfico de correlação com outra configuração, para melhor visualização.'''
'''Dessa forma, quando mais azul o quadrante, mais próximo de uma correlação positiva.'''
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(
    corr,
    annot = True,
    mask = mask,
    square = True,
    cmap = 'Blues');
```
